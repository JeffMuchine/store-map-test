```python
import numpy as np
input = '6682json.txt'
x = np.loadtxt(input, dtype=str,delimiter=']],')
x[-1] = x[-1][:-2]
```

```python
f = open('text.txt','w')
for item in x:
    a = np.array(item.split(':[['),dtype=str)
    b = np.array(a[1].split(','), dtype = str)
    text = '<text x="%s" y="%s">%s</text>'%(b[0], b[1], a[0])
    f.write('%s\n'%text)
f.close()
```
